# Exercise 1.04: Applying Data Transformations

```python
df = pd.read_json("../Chapter01/Exercise1.04/user_info.json")
df.head()
```

```python
df.info()
```

```python
# What is the average age of the users?
df["age"].mean()
```

```python
# Which is the favorite fruit of the users?
df["favoriteFruit"].max()
```

```python
# Which is the favorite fruit of the users?
df.groupby("favoriteFruit")["_id"].max()
```

```python
# Do you have more female customers?
len(df[df["gender"] == "female"])
```

```python
# Do you have more female customers?
df.groupby("gender")["_id"].count()
```

```python
# How many of the users are active?
len(df[df["isActive"] == True])
```

```python
# How many of the users are active?
df.groupby("isActive")["_id"].count()
```

# Activity 1.01: Addressing Data Spilling

```python
import pandas as pd
```

```python
df = pd.read_csv("../Chapter01/Activity1.01/sales.csv")
df
```

```python
df.info()
```

```python
df["Product line"] = df["Product"] + " " + df["line"]
df = df.drop(["Product", "line"], axis = 1)
df
```

```python
df["Product Type"] = df["Product.1"] + df["type"]
df = df.drop(["Product.1", "type"], axis = 1)
df
```

```python
df["Product Order Method"] = df["Product.2"] + " " + df["Order"] + " "+ df["method"]
df = df.drop(["Product.2", "Order", "method"], axis = 1)
df
```

```python
df.groupby("type.1")["Year"].count()
```

```python
df = df.rename(columns = {
    "type.1" : "Order Type"
})
df
```

```python
df = df.fillna("")
```

```python
df["Retailer Country"] = df["Retailer"] + " " + df["country"]
df = df.drop(["Retailer", "country"], axis = 1)
```
