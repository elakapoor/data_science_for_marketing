# Exercise 2.02: Calculating Conversion Ratios for Website Ads.

```python
df = pd.read_csv("./Exercise2.02/conversion_rates.csv")
df

# If an advertisement succeeds in getting a user to click on it, the converted field gets the value 1, 
# otherwise, it gets 0 by default; the group field denotes which ad was clicked – A or B.
```

```python
# how many people have viewed the advertisement
# Store the result in a DataFrame viewed_df. Also, make sure you change the column name from converted to viewed
view_df = df.groupby("group").count().rename({"converted":"viewed"}, axis = "columns")
view_df
```

```python
# count the number of conversions.
converted_df = df.loc[df["converted"] == 1].groupby("group")["converted"].count()
converted_df
```

```python
converted_df = df.groupby("group").sum()
converted_df
```

```python
# Combine the converted_df and viewed_df datasets in a new DataFrame, named stats 
stats = pd.merge(converted_df,view_df, on = "group" )
stats
```

```python
# Create a new column called conversion_ratio that displays the ratio of converted ads to the number of views 

stats["conversion_ratio"] = stats["converted"]/stats["viewed"]
stats
```

```python
df_1 = stats.stack()
df_1
```

```python
df_1["A"]["conversion_ratio"]
```

```python
stats.unstack().unstack()
```

```python
# You can use the pivot function to change the columns to rows as follows:
df.pivot(columns = "group", values = "converted")
```
